```python
# Function to classify sentiment
def classify_sentiment(content):
    scores = sid.polarity_scores(content)
    if scores['compound'] >= 0.05:
      return 'positive'
        #print( 'positive')
    elif scores['compound'] <= -0.05:
      return 'negative'
        #print( 'negative')
    else:
        return 'neutral'
```

```python
# Tokenize and remove stopwords
stop_words = stopwords.words('english') # Use a list instead of a set
vectorizer = CountVectorizer(stop_words=stop_words)
```

```python
def get_most_common_words(contents):
    X = vectorizer.fit_transform(contents)
    word_counts = X.sum(axis=0).A1
    words = vectorizer.get_feature_names_out()
    word_freq = dict(zip(words, word_counts))
    sorted_words = sorted(word_freq.items(), key=lambda item: item[1], reverse=True)
    return sorted_words[:10]
```

```python
# Get most common words for each sentiment
positive_contents = bank_df[bank_df['sentiment'] == 'positive']['content']
negative_contents = bank_df[bank_df['sentiment'] == 'negative']['content']
neutral_contents = bank_df[bank_df['sentiment'] == 'neutral']['content']
```

```python
most_common_positive = get_most_common_words(positive_contents)
most_common_negative = get_most_common_words(negative_contents)
most_common_neutral = get_most_common_words(neutral_contents)
```

```python
print("Most common words in positive reviews:", most_common_positive)
print("Most common words in negative reviews:", most_common_negative)
print("Most common words in neutral reviews:", most_common_neutral)
```

```python
# Ensure all entries in 'content' are strings and filter out empty reviews
bank_df['content'] = bank_df['content'].astype(str)
bank_df = bank_df[bank_df['content'].str.strip().astype(bool)]
```

```python
# Initialize sentiment analyzer
sid = SentimentIntensityAnalyzer()
```

```python
!pip install emoji
import emoji

# Function to convert emojis to text
def convert_emojis(text):
    return emoji.demojize(text) # Call demojize from the emoji module
```

```python
def classify_sentiment(content):
    content = convert_emojis(content)
    scores = sid.polarity_scores(content)
    if scores['compound'] >= 0.05:
        return 'positive'
    elif scores['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'
```

```python
# Apply sentiment analysis
bank_df['sentiment'] = bank_df['content'].apply(classify_sentiment)
```

```python
# Print the results
for index, row in bank_df.iterrows():
    print(f"Review: {row['content']}\nSentiment: {row['sentiment']}\n")
```
